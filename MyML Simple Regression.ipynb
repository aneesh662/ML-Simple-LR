<h1><center> ML Simple Linear Regression </h1>

<h2>Step 1: Import Necessary Libraries</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

<h2>Step 2: Import the Dataset</h2>

```python
# Load the dataset
data = pd.read_csv(r"Experience Salary.csv")

# Display the dataset
print(data)
```

<h2>Step 3: Prepare the Data</h2>

```python
#Split the dataset into features (X) and target (y), then further split into training and testing sets.
# Define features and target variable
X = data[['Experience']]
y = data['Salary']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

<h2>Step 4: Train the Model</h2>

```python
# Create a Linear Regression object
model = LinearRegression()

# Fit the model with the training data
model.fit(X_train, y_train)
```

<h2>Step 5: Make Predictions>/h2>

```python
# Predict on the test set
y_pred = model.predict(X_test)

# Predict a specific value
predicted_salary_for_5_years = model.predict([[5]])
predicted_salary_for_12_years = model.predict([[12]])

print(f"Predicted salary for 5 years of experience: {predicted_salary_for_5_years[0]}")
print(f"Predicted salary for 12 years of experience: {predicted_salary_for_12_years[0]}")
```

<h2>Step 6: Evaluate the Model</h2>

```python
#Check the accuracy, coefficient, and intercept of the model.

# Model accuracy
accuracy = model.score(X_test, y_test)
print(f"Model accuracy: {accuracy}")

# Model coefficients
print(f"Model coefficient (m): {model.coef_[0]}")
print(f"Model intercept (b): {model.intercept_}")

# Calculate and print the Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

<h2>Step 7: Visualize the Results</h2>

```python
# Plot the data points and the regression line
plt.scatter(data['Experience'], data['Salary'], color='r', marker='.')
plt.plot(data['Experience'], model.predict(data[['Experience']]), color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()
```
